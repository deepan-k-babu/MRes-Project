<h2 style="font-family: 'Times New Roman'; font-weight: bold;">
    Preprocessing
</h2>

```python
# Remove the column header and make the column 1 as column heading
lignin.columns = lignin.iloc[0]
lignin = lignin[1:]
lignin
```

```python
# Drop null values
lignin = lignin.dropna(axis=1, how='all')
lignin
```

```python
# Remove Duplicate Columns
lignin = lignin.loc[:, ~lignin.columns.duplicated()]
lignin
```

```python
# Removing the last two letters of sample values
lignin['IRRI seed bank ID'] = lignin['Sample'].str[:-2]
lignin
```

```python
# Creating a column with IRRI seed bank ID from Sample ID
lignin['IRRI seed bank ID'] = lignin['IRRI seed bank ID'].str[:4]+' '+lignin['IRRI seed bank ID'].str[4:]
lignin
```

```python
lignin['IRRI seed bank ID'].unique()
```

```python
# Removing the extra '_' in the IRRI seed bank ID column
lignin['IRRI seed bank ID'] = lignin['IRRI seed bank ID'].str.rstrip('_')
lignin['IRRI seed bank ID'].unique()
```

```python
# Preprocessing seed cultivar data
supdata1.columns = supdata1.iloc[0]
supdata1 = supdata1[1:]
supdata1
```

```python
# matching IRRI seed bank ID column of both data frame
lignin['IRRI seed bank ID'].nunique() == supdata1['IRRI seed bank ID'].nunique()
```

```python
len(set(supdata1['IRRI seed bank ID'].unique()).intersection(set(lignin['IRRI seed bank ID'])))
```

```python
nunique = supdata1.groupby('IRRI seed bank ID')['SRA accession'].nunique()
print(nunique)
```

```python
# creating a df with IRRI seed bank ID and MAGs' SRA accession to map the source of MAGs
unique = supdata1.groupby('IRRI seed bank ID')['SRA accession'].unique()

match = {
    'IRRI seed bank ID': [],
    'sra': []
}

for key, values in unique.items():
    match['IRRI seed bank ID'].append(key)
    match['sra'].append('; '.join(map(str, values)))

match = pd.DataFrame(match)

match
```

```python
lignin
```

```python
(lignin.groupby('IRRI seed bank ID')['Subgroup'].nunique() == 1).unique()
```

```python
# drop the sample column from lignin df
lignin = lignin.drop(columns=['Sample'])
lignin
```

```python
# Take the average of technical replicates of lignin precursor concentration.
avg_lignin = lignin.groupby('IRRI seed bank ID').agg({
    'Subgroup': lambda x: x.iloc[0] if x.nunique() == 1 else '; '.join(x),
    'HCA (mg/g)': lambda x: x.sum()/10,
    'LPA (mg/g)': lambda x: x.sum()/10,
    'FLA (mg/g)': lambda x: x.sum()/10,
    'CFA (mg/g)': lambda x: x.sum()/10,
    'SPLT (mg/g)': lambda x: x.sum()/10,
}).reset_index()

avg_lignin
```

```python
# Merge the SRA accession data df and processed lignin precursor df
avg_lignin = pd.merge(avg_lignin, match, on='IRRI seed bank ID', how='inner')
avg_lignin
```

```python
# Merge with coverm preparation df
avg_lignin['Genomes'] = ''
for i, row in avg_lignin.iterrows():
    values = row['sra'].split('; ')
    matching_values = [str(coverm.loc[coverm['SRA accession'] == value, 'New SRA'].values[0]) for value in values if not coverm.loc[coverm['SRA accession'] == value, 'New SRA'].empty]
    avg_lignin.at[i, 'Genomes'] = '; '.join(matching_values)

avg_lignin
```

```python
# removing the unassigned rice cultivar data from df
avg_lignin = avg_lignin[~avg_lignin['IRRI seed bank ID'].isin(unassigned)]
avg_lignin
```

```python
avg_lignin = avg_lignin.drop(columns=['sra'])
avg_lignin
```

```python
# save the genome grouping df as .csv file
genomes = []

for index, row in avg_lignin.iterrows():
    genome = row['Genomes'].split('; ')
    for values in genome:
        genomes.append({'Metagenomes': values, 'Group': row['Subgroup']})

grouping = pd.DataFrame(genomes)

grouping.to_csv('C:\\Users\\dkb1g24\\Documents\\mres\\raw_data\\groups.csv', index=False)
```
